```python
#collecting required columns to build the model. 
# As we know Customer ID, Age, and Gender does 
#not making much contribution in classinfying the segmentation we will take only columns 3 and 4
X=customer_data.iloc[:,[1,2,3,4,5,6]]
#X.dropna()
```

```python
wcss=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init="k-means++",random_state=42)
    kmeans.fit(X)
    
    wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow point graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
```

```python
kmeans=KMeans(n_clusters=4, init="k-means++",random_state=0)
Y=kmeans.fit_predict(X)
print(Y)
```

```python
data_with_cluster=pd.concat([customer_data,pd.DataFrame({'Cluster': kmeans.labels_})],axis=1)
data_with_cluster.head()
```

```python
sns.countplot(x='Cluster', data=data_with_cluster)
```
